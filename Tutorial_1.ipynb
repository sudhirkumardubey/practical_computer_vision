```python
# get dimensions of image
dimension = img.shape
print('Image dimension: ', dimension)

if len(dimension) == 2:
    # if image is grayscale
    print("Grayscale image:")
    height = img.shape[0]
    weight = img.shape[1]
    print('Image size: ', height, 'x', weight)

if len(dimension) == 3:
    # if image is color
    print("Color image:")
    height = img.shape[0]
    weight = img.shape[1]
    channel = img.shape[2]
    print('Image size: ', height, 'x', weight, 'x', channel)

# dtype of image
print('Image dtype: ', img.dtype)
```

```python
# Creating empty image with width and height of the src image
zeroImg = np.zeros((img.shape[0], img.shape[1]), img.dtype)

# # Copying the src image into the empty image
# zeroImg[:img.shape[0], :img.shape[1], :img.shape[2]] = img # Alternative way 

# get channels
(red_channel, green_channel, blue_channel) = cv2.split(img)

# create new images with three channels per image and assign color channels to new channels and set remaining channels to zero zero iamges
red_img = cv2.merge([red_channel, zeroImg, zeroImg])
green_img = cv2.merge([zeroImg, green_channel, zeroImg])
blue_img = cv2.merge([zeroImg, zeroImg, blue_channel])
```

```python
# Iterate by loop and replacing pixel with the desired values
img = cv2.imread('Image/pier.jpg',1)

print(img.shape)

if len(img.shape) ==2:
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            if img[y,x] == 125: 
                img[y,x] = 255

if len(img.shape)==3:
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            for c in range(img.shape[2]):
                if img[y,x,c] == 125: 
                    img[y,x,c] = 255

# show image
plt.imshow(img)
plt.show()
```

```python
# Thresholding

img = cv2.imread('Image/pier.jpg', 1)

Threshold = 155

if len(img.shape) == 2:
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            img[y, x] = 255 if img[y,x] >=Threshold else 0

if len(img.shape)==3:
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            for c in range(img.shape[2]):
                img[y, x, c] = 255 if img[y,x,c] >=Threshold else 0

# show image
plt.imshow(img)
plt.show() 
   
```

```python
img = cv2.imread('Image/rgb.jpg',0)
plt.imshow(img)
plt.show()

# To invert pixel value

if len(img.shape) == 2:
    new_img = np.zeros((img.shape[0], img.shape[1]), img.dtype)
    max = 0

    # Loop to get maximum pixel value
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i, j] > max:
                max = img[i,j]
    
    # Loop to inverse the image
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            new_img[i,j] = max-img[i,j]
        

print("Max pixel: ", max)
plt.imshow(new_img)
plt.show()
```
